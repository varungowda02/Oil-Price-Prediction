```python
def replace_outliers_with_rolling_median(series, window=7, threshold_factor=3):
    rolling_median = series.rolling(window=window, center=True).median()
    mad = (series - rolling_median).abs().rolling(window=window, center=True).median()
    threshold = threshold_factor * mad

    # Mask outliers
    mask = (series - rolling_median).abs() > threshold

    # Copy series and replace only outliers
    new_series = series.copy()
    new_series[mask] = rolling_median[mask]

    return new_series

# Apply only to numeric columns
for col in df.select_dtypes(include='number').columns:
    df[col] = replace_outliers_with_rolling_median(df[col], window=7, threshold_factor=3)

# Interpolate any NaN created at edges
df.interpolate(method='linear', inplace=True)
```

```python
df['Volume'] = replace_outliers_with_rolling_median(df['Volume'], window=45, threshold_factor=3)
```

A longer window (45 days) smooths short-term volatility and catches broader market patterns.

threshold_factor=3 makes the method stricter, reducing extreme spikes without over-smoothing.

We are treating outliers, not the whole series, so genuine trends stay intact.

```python
# probability of crude oil daily price changes falling within 1%, 2%, 5%, and 10%.

# Daily percent changes
df2_pct = df2['Close/Last'].pct_change().dropna()

# Define thresholds
thresholds = [0.01, 0.02, 0.05, 0.10]

# Store probabilities
probs = {}
for t in thresholds:
    probs[f"Within {int(t*100)}%"] = (df2_pct.between(-t, t).mean() * 100)

# Display results
print("Probabilities of Daily Percentage Changes in Crude Oil Prices:")
for k, v in probs.items():
    print(f"{k}: {v:.2f}%")
```

Within 2%: 77.70%

Within 5%: 95.96%
 About 96 out of 100 days the price changes by less than 5%.

Within 10%: 99.21%

```python
import warnings
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import itertools

# Set the parameters range
p = range(0, 4)
d = range(0, 2)
q = range(0, 4)

# Generate all possible combinations of p, d, and q
pdq_combinations = list(itertools.product(p, d, q))

# Suppress warnings
warnings.filterwarnings("ignore")

# Grid search for the best ARIMA model
best_score, best_pdq = float("inf"), None
for param in pdq_combinations:
    try:
        model = ARIMA(df['Close/Last'], order=param)
        model_fit = model.fit()
        predictions = model_fit.predict(start=len(df)-30, end=len(df)-1)
        error = mean_squared_error(df['Close/Last'][-30:], predictions)
        if error < best_score:
            best_score = error
            best_pdq = param
        print(f'ARIMA{param} MSE={error}')
    except:
        continue

print(f'Best ARIMA parameters: {best_pdq} with MSE={best_score}')
```

```python
# Import necessary libraries
import itertools
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Define parameter ranges for alpha, beta, and gamma
alpha_values = [0.1, 0.2, 0.5, 0.8]   # Level smoothing
beta_values = [0.1, 0.2, 0.5, 0.8]    # Trend smoothing
gamma_values = [0.1, 0.2, 0.5, 0.8]   # Seasonal smoothing

# Define the trend and seasonal type, here we applied additive method
trend_type = 'add'     # 'add' for additive trend, 'mul' for multiplicative
seasonal_type = 'add'  # 'add' for additive seasonality, 'mul' for multiplicative
seasonal_periods = 12  # Number of periods in the seasonal cycle (monthly data)
```

```python
# Function to perform grid search for Exponential Smoothing
def optimize_exponential_smoothing(df, alpha_values, beta_values, gamma_values, trend, seasonal, seasonal_periods):
    best_aic = float("inf")
    best_model = None
    best_params = None

    # Generate all combinations of alpha, beta, gamma
    param_combinations = itertools.product(alpha_values, beta_values, gamma_values)

    for alpha, beta, gamma in param_combinations:
        try:
            # Fit the Exponential Smoothing model with current parameters
            model = ExponentialSmoothing(df['Close/Last'],
                                         trend=trend,
                                         seasonal=seasonal,
                                         seasonal_periods=seasonal_periods)

            fitted_model = model.fit(smoothing_level=alpha, smoothing_slope=beta, smoothing_seasonal=gamma)

            # Calculate AIC and check if it is the best model
            aic = fitted_model.aic
            print(f'alpha={alpha}, beta={beta}, gamma={gamma}, AIC={aic}')

            if aic < best_aic:
                best_aic = aic
                best_model = fitted_model
                best_params = (alpha, beta, gamma)

        except Exception as e:
            print(f"Model failed for alpha={alpha}, beta={beta}, gamma={gamma} due to {e}")
            continue

    print(f'\nBest AIC: {best_aic}')
    print(f'Best Parameters: alpha={best_params[0]}, beta={best_params[1]}, gamma={best_params[2]}')
    return best_model

# Perform the grid search
best_model = optimize_exponential_smoothing(df, alpha_values, beta_values, gamma_values, trend_type, seasonal_type, seasonal_periods)
```

```python
# Define the train-test split ratio (e.g., 80% train, 20% test)
train_size = int(len(df) * 0.8)
test_size = len(df) - train_size

# Split the data into training and testing sets
train, test = df.iloc[:train_size], df.iloc[train_size:]

# Display the shapes of train and test sets
print(f'Train set shape: {train.shape}')
print(f'Test set shape: {test.shape}')
```

```python
# Forecast the test period (same length as the test set)
forecast = best_model.forecast(steps=len(test))

# Create a DataFrame for the forecasted values
forecast_df = pd.DataFrame(forecast, index=test.index, columns=['Forecast'])

# Plot actual vs forecasted values
plt.figure(figsize=(10,6))
plt.plot(train['Date'], train['Close/Last'], label='Train')
plt.plot(test['Date'], test['Close/Last'], label='Test')
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecast', color='green')
plt.title('Exponential Smoothing Model Forecast vs Actual Values')
plt.xlabel('Date')
plt.ylabel('Close/Last')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a DataFrame of the predictor variables
X = train_data[['High', 'Low', 'Open', 'Volume']]

# Calculate VIF for each predictor
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("VIF values for predictor variables:")
display(vif_data)

# Interpretation of VIF values:
# VIF > 10 indicates high multicollinearity (some sources say > 5)
# If VIF is high, consider removing that predictor or combining with others.

# Suggestion based on VIF
high_vif_features = vif_data[vif_data['VIF'] > 10]['feature'].tolist() # Using 10 as a common threshold

if high_vif_features:
    print("\nPredictors with high VIF (indicating multicollinearity):")
    for feature in high_vif_features:
        print(f"- {feature}")
    print("\nConsider dropping or transforming these features to reduce multicollinearity.")
else:
    print("\nNo predictors with high VIF values (VIF <= 10). Multicollinearity does not appear to be a significant issue based on this test.")
```
